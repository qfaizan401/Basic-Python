```python
#def cls():
#    print('\n'*50)
#cls()
from colorama import Fore, Back, Style
school_name='ABC SCHOOL'.center(110)
massage1='Student Result Card'.center(110)
#print('\n'*2)
print(Fore.BLACK,
     Back.BLUE,
     Style.BRIGHT,
     school_name,
     Style.RESET_ALL)
print('\n'*0)
print(Fore.BLACK,
     Back.BLUE,
     Style.BRIGHT,
     massage1,
     Style.RESET_ALL)
student_name=input('\n'"Enter the Student Name:"'\t\t\t\t')
sub1=float(input('\n'"Enter the marks of Subject 1:"'\t\t\t'))
sub2=float(input('\n'"Enter the marks of Subject 2:"'\t\t\t'))
sub3=float(input('\n'"Enter the marks of Subject 3:"'\t\t\t'))
sub4=float(input('\n'"Enter the marks of Subject 4:"'\t\t\t'))
sub5=float(input('\n'"Enter the marks of Subject 5:"'\t\t\t'))
sub6=float(input('\n'"Enter the marks of Subject 6:"'\t\t\t'))
print('\n')
if sub1<50.0:
    print(Fore.BLACK,
         Back.RED,
         Style.BRIGHT,
         f"You have FAILED in Suject 1 with obtain marks:{sub1}",
         Style.RESET_ALL)
if sub2<50.0:
    print(Fore.BLACK,
         Back.RED,
         Style.BRIGHT,
         f"You have FAILED in Suject 2 with obtain marks:{sub2}",
         Style.RESET_ALL)
if sub3<50.0:
    print(Fore.BLACK,
         Back.RED,
         Style.BRIGHT,
         f"You have FAILED in Suject 3 with obtain marks:{sub3}",
         Style.RESET_ALL)
if sub4<50.0:
    print(Fore.BLACK,
         Back.RED,
         Style.BRIGHT,
         f"You have FAILED in Suject 4 with obtain marks:{sub4}",
         Style.RESET_ALL)
if sub5<50.0:
    print(Fore.BLACK,
         Back.RED,
         Style.BRIGHT,
         f"You have FAILED in Suject 5 with obtain marks:{sub5}",
         Style.RESET_ALL)
if sub6<50.0:
    print(Fore.BLACK,
         Back.RED,
         Style.BRIGHT,
         f"You have FAILED in Suject 6 with obtain marks:{sub6}",
         Style.RESET_ALL)
if sub1>50.0 and sub2>50.0 and sub3>50.0 and sub4>50.0 and sub5>50.0 and sub6>50.0:
    print(Fore.BLACK,
         Back.GREEN,
         Style.BRIGHT,
         "You have Passed all the subjects".center(110),
         Style.RESET_ALL)
total_marks=sub1+sub2+sub3+sub4+sub5+sub6
print('\n'f"TOTAL MARKS: {total_marks}")
per=round((total_marks/600)*100,2)
print(f"PERCENTAGE: {per}")
if per>=90.0:
    print('\n',Fore.BLACK,
         Back.GREEN,
         Style.BRIGHT,
         "Grade: A+".center(110),
         Style.RESET_ALL)
elif 80.0<per<90.0:
    print('\n',Fore.BLACK,
         Back.LIGHTGREEN_EX,
         Style.BRIGHT,
         "Grade: A".center(110),
         Style.RESET_ALL)
elif 70.0<per<80.0:
    print('\n',Fore.BLACK,
         Back.YELLOW,
         Style.BRIGHT,
         "Grade: B".center(110),
         Style.RESET_ALL)
elif 60.0<per<70.0:
    print('\n',Fore.BLACK,
         Back.CYAN,
         Style.BRIGHT,
         "Grade: C".center(110),
         Style.RESET_ALL)
elif 50.0<per<60.0:
    print('\n',Fore.BLACK,
         Back.MAGENTA,
         Style.BRIGHT,
         "Grade: D".center(110),
         Style.RESET_ALL)
else:
    print('\n',
          Fore.BLACK,
         Back.RED,
         Style.BRIGHT,
         "FAILED".center(110),
         Style.RESET_ALL)
```
